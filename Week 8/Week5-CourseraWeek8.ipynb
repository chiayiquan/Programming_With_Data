### Absolute vs Relative Path 


* The absolute path is the full path starting from the root or C drive to a given file or directory.
* The relative path is the path to a given file or directory, with respect to the current working directory `(cwd)`. For example:
  ** Absolute path: C:/users/julian/docs/myfile.txt
  ** If my cwd is C:/users/julian/, then the relative path to `myfile.txt` would be: `docs/myfile.txt`

That is to say, `cwd` + relative path = absolute path.
