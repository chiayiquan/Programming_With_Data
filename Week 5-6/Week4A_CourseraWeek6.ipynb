## Definition:
### Term Frequency
TF is defined as the ratio of a words occurrence in a document to the total number of words in a document

### Inverse Document Frequency (IDF)

IDF `reduces` the weight of terms that `occur frequently`, and `increases` the weight of terms that `occur rarely`.

Why?<br>
Because Term frequency (TF) does not consider the importance of words. Some words such as` of, and, etc.` appear the most, but are of little significance. ####

* Manually Calculate Term Frequency
* Manually Calculate Inverse Document Frequency



````python
import functools

first_sentence_list = first_sentence.lower().split(' ')
second_sentence_list = second_sentence.lower().split(' ')
first_sentence_len = len(first_sentence_list)
second_sentence_len = len(second_sentence_list)
print(first_sentence_len)
print(second_sentence_len)
````

````python
def count(accumulator,currentValue):
    if currentValue in accumulator:
        return {**accumulator, currentValue:accumulator[currentValue]+1}
    else:
        return {**accumulator,currentValue:1}
first_sentence_freq = functools.reduce(count,first_sentence_list,{})
second_sentence_freq = functools.reduce(count,second_sentence_list,{})
print(first_sentence_freq)
print(second_sentence_freq)
````

````python
first_sentence_idf = [2/first_sentence_freq[freq] for freq in first_sentence_freq]
print(first_sentence_idf)
````
