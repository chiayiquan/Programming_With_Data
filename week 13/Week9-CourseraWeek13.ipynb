#### Based on the mean of result

```python
import pandas as pd
import numpy as np
from numpy import nan as NaN
students = [[19, 86, 'M', 'verygood'],
           [21, 96, 'F', 'excellent'],
           [32, 75, None,'good'],
           [np.NaN, np.NaN, 'M', 'good'],
           [np.NaN, 71, 'M', 'good'],
           [25, np.NaN, None, 'verygood'],
           [32, 92, 'F', 'verygood'],
           [np.NaN, 98, 'M', 'excellent']]
df = pd.DataFrame(students)
df.columns = ['age', 'marks', 'gender', 'result']
```

```python
df
```

```python
goodmean= df[df.result=='good'].marks.mean()
```

```python
goodmean
```

```python
verygoodmean= df[df.result=='verygood'].marks.mean()
```

```python
verygoodmean
```

```python
for i in df[df.marks.isna()].index:
    print(df.iloc[i, 3])
    if df.iloc[i, 3] == 'good':
        df.iloc[i, 1] = goodmean
    if df.iloc[i, 3] == 'verygood':
        df.iloc[i, 1] = verygoodmean
        
```
