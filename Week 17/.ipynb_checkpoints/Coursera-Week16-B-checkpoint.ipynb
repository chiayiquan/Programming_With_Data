# 8.105 Intro to test-driven development

### What is a unit test?

A unit test is 
* a piece of code commonly used to test the functionality of a unit.
* It is written and run by `software developers` 
* To ensure that a unit meets its design and behaves properly.
* A unit is commonly an individual function. 


### Types of Functional Testing
1. Unit Testing
1. Integration Testing:
   * Testing of all integrated function, checking whether the combined functions work properly after integration
1.System Testing:
   * An entire system is tested to see whether it fulfils the specifications 
   * It is a Black-box type Testing.
1. Alpha/Beta/Acceptance Testing:
   * Alpha Testing is carried out by software development or internal employees. 
   * Beta Testing is a formal type of Software Testing which is carried out by the customer.
   * Acceptance or User Acceptance Testing(UAT) is the last phase of the testing. The software goes into production after passing this test

## A. Theory of test driven development

### * Who invented test-driven development <br>  * Definition of test-driven development <br>  * Three laws of test-driven development




## B. Who invented test-driven development
#### * Beck (1999): Extreme Programming (Quick Iterated Cycles) <br> * Beck K, Andres C (2004) Extreme Programming Explained: Embrace Change (2nd Edition). Addison-Wesley Professional




## C. Definition of test-driven development
#### Test Driven Development (TDD) is a practice advocated by agile software development methodologies, in which 
#### * `tests are written in advance` of source-code development. These tests, intended to <br> * `initially fail in the absence` of any substantial implementation, effectively <br> * `constitute a specification of the functionality and behavior of the software `code, which can be tested as it is being developed (Beck 2003).  i.e.,  the set of unit test is really a specification for the code





Neil C. Borle1  Meysam Feghhi1 Eleni Stroulia1 Russell Greiner1,2  Abram Hindle1
Analyzing the effects of test driven development in GitHub, Empir Software Eng
https://doi.org/10.1007/s10664-017-9576-3


 
## D. The Three Laws of test-driven development by R. C. Martin

R.C Martins, "Professionalism and Test-Driven Development," in IEEE Software (24:3), pp.33-36.

### * 1. You may not write production code `unless you first written a failing unit test`.  <br> * 2. you may not write `more of a unit test` that is sufficient to fail (write minimum code to fail). <br> * 3. you `may not write more production code` of any sufficient to make the failing unit test `pass` (write minimun code to pass the test i.e., production).


### The three phases of Test Driven Development? <br> * Red, Green, Refactor Test: <br> * In the Red phase, you implement a failing unit test. <br> * in the Green phase, you make it pass <br> * Lastly, in the refactor phase you modify the production code and the test to remove duplication and make the code clean.

### Function

```python
def SaGo(input):
    if input%3==0:
        return 'Sa'
    elif input%5==0:
        return 'Go'
    else:
        return input
```

### Test Cases

```python
import unittest


class Test_UseCase1(unittest.TestCase):
    
    def test_SaGo_exist(self):
        self.assertIsNotNone(SaGo)
        
    def test_SaGo_1(self):
        self.assertEqual(SaGo(1),1)    
        
    def test_SaGo_2(self):
        self.assertEqual(SaGo(2),2)    

                
    def test_SaGo_3(self):
        self.assertEqual(SaGo(3),'Sa')    
                              
    def test_SaGo_4(self):
        self.assertEqual(SaGo(4),4)    
                        
    def test_SaGo_5(self):
        self.assertEqual(SaGo(5),'Go')    
                        
    def test_SaGo_6(self):
        self.assertEqual(SaGo(6),'Sa')    

    def test_SaGo_10(self):
        self.assertEqual(SaGo(10),'Go')    
   
        
unittest.main(argv=['ignored', '-v'], exit=False)
    
```

Below can be found on wiki:

Limitations and disadvantages
Testing will not catch every error in the program, because it cannot evaluate every execution path in any but the most trivial programs. This problem is a superset of the halting problem, which is undecidable. The same is true for unit testing. Additionally, unit testing by definition only tests the functionality of the units themselves. Therefore, it will not catch integration errors or broader system-level errors (such as functions performed across multiple units, or non-functional test areas such as performance). Unit testing should be done in conjunction with other software testing activities, as they can only show the presence or absence of particular errors; they cannot prove a complete absence of errors. To guarantee correct behavior for every execution path and every possible input, and ensure the absence of errors, other techniques are required, namely the application of formal methods to proving that a software component has no unexpected behavior.[citation needed]
