**Relational Database**

Assume we have two separate tables that have information about the `customer` and the `order`. We might then import the `customer's ID` or `userid` into the order table when an order is made. We will also have some `unique identifier` in the customer table to tell us that this is a mapping to a specific customer (e.g, customer's address, phone, payment)

Every time that customer makes an order, we `can take `that information and import it into our order table to say it's this particular customer that's making an order. In some instances, we might have one customer `having one or many orders`, so they may have just ordered once or they may have ordered a 100 times. It's important that we can maintain these records and make sense of them and keep them.
 


**Schema**<br>
We might want to implement a schema that `informs entity relationships`, that `defines` what these relationships look like between these different tables. 

Two kinds of database schema:
* A `logical database schem`a conveys the logical constraints that apply to the stored data. It may define integrity constraints, views, and tables.
* A `physical database schema` lays out how data is stored physically on a storage system in terms of files and indices.


One database can have many tables, and can have many relationship within the table. Therefore, one database can have many schemas, not the other way round.

note:
`Integrity` constraints are a set of rules, used to maintain the `quality of information`(i.e., data integrity), and ensure that the data insertion, updating, and other processes have to be performed in such a way that data integrity is not affected.

## Entity Relationship Diagram

An Entity Relationship Diagram (ERD) is a visual representation of different entities, relationships, and attributes within a system and how they relate to each other

### The use of ERD:
* To define terms related to entity relationship modeling
* To provide information about all tables, and how they should connect, what fields are within  each table
* To Help to describe entities, attributes, relationships
* ER diagrams can be converted into relational tables, we can build databases quickly

### Three basic elements in ERD:
1. Entity
2. Attribute
3. Relationship

###  Entity:
An entity can be place, person, object, event or a concept.  The entities  must have an attribute, and a primary (unique) key.  Examples are Student, Class, Teacher, Patient, Store, Order, Car and so on.

###  Attribute:
An attribute is a property, trait, or characteristic of an entity, relationship, or another attribute. It is a single-valued property of either an entity-type or a relationship-type.

 Attributes in ER diagrams, Note that an attribute can have its own attributes ( composite attribute ) https://creately.com/blog/diagrams/er-diagrams-tutorial/


###  Relationship:
A relationship describes how entities interact. Relationship is an association among two or more entities. For example, the entity “Carpenter” may be related to the entity “table” by the relationship “builds” or “makes”.
 Using Relationships in Entity Relationship Diagrams


### Cardinality:
Defines the relationship between two entities by placing number
Different types of cardinal relationships are:
* One-to-One Relationships
* One-to-Many Relationships
* Many to One Relationships
* Many-to-Many Relationships



 A student can be assigned multiple courses and a Professor can deliver only one course https://www.guru99.com/er-diagram-tutorial-dbms.html


### Primary Key:
After having a list of Attributes, we need to specify the identified entitie, ensuringe an attribute is to be paired with exactly one entity. The attribute is called Primary Key.

If you are not sure, take a look at a few examples online first. There are quite a few acceptable ways to design ERDs so do not despair if you see a different visual structure.
