```python
# Load r&j text
with open("romeo-and-juliet_TXT_FolgerShakespeare.txt") as f:
    data=f.read()
```

```python
# display a pie chart
def displayPieChart(display_data,title):
    fig, ax = plt.subplots()
    ax.pie(list(map(lambda word:word['value'],display_data)), labels=list(map(lambda word:word['label'],display_data)), autopct='%1.2f%%')
    ax.axis('equal')  
    ax.set_title(title)
    plt.show()
```

```python
# display a table
def displayPDTable(display_data):
    data_frame=pd.DataFrame({"appearance":list(map(lambda word:word[1],display_data))},index=list(map(lambda word:word[0],display_data)))
    return print(data_frame)
```

```python
# get the first 10 words
def firstTenWords(word):
    return [(key,word[key]) for key in list(word)[:10]]
```

```python
# generate word cloud
def generateWordCloud(words):
    wcloud=WordCloud().generate_from_frequencies(words)
    wcloud.background_color='black'
    plt.imshow(wcloud, interpolation='bilinear')
    plt.axis("off")
    (-1.5, 200, 100, -2.5)
    plt.show()
```

```python
# take in a dict
def sortDictByValue(dict_):
    return {key: value for key, value in sorted(dict_.items(), key=lambda item: item[1],reverse=True)}
     
```

```python
# tokenize the word
split_words=nltk.word_tokenize(data) # why not use string.split(' ') because words like Verona, will be counted as 1 where tokenize will split Verona and , into individual 
lower_case_words=[word.lower() for word in split_words] # convert all word to lowercase to prevent duplicated words

print(lower_case_words)
```

```python
total_character=len(lower_case_words)
print(f'There are a total of {total_character} words/character when we tokenize this romeo and juliet text.')
```

```python
# count the word freq
data_analysis = nltk.FreqDist(lower_case_words)
data_analysis
```

```python
# set stop words as english
stop_words = set(stopwords.words('english')) 

# Add in stop words of meaningless characters
my_lists={',','.','the','i','and','to','a','of','is','============','[',']','(',')','!','?','--',';','\'','`','\'s'}
stop_words.update(my_lists)
```

```python
# remove all stop words
filtered_words = [word for word in lower_case_words if not word in stop_words] 
filtered_words_len=len(filtered_words)

displayPieChart([{'label':'Filler','value':total_character-filtered_words_len},{'label':'Meaningful Words','value':filtered_words_len}],'Words Percentage')
```

```python
# count the word freq of the filtered list
freq_filtered_word = nltk.FreqDist(filtered_words)
```

```python
# lemmatize the word in verbs
lemmatizer = WordNetLemmatizer()
lemmatized_word = list(map(lambda word:lemmatizer.lemmatize(word, pos="v"),filtered_words))
```

```python
freq_lemmatized_word= nltk.FreqDist(lemmatized_word)
```

```python
# get the top 10 words of not lemmatize word
top_ten_word_non_lemmatized = firstTenWords(freq_filtered_word)
```

```python
# get the top 10 words of lemmatized word
top_ten_word_lemmatized = firstTenWords(freq_lemmatized_word)
```

### Top 10 non lemmatize word

```python
# Non lemmatize word data frame
displayPDTable(top_ten_word_non_lemmatized)
generateWordCloud(freq_filtered_word)
```

### Top 10 lemmatized word

```python
# Lemmatized word data frame
displayPDTable(top_ten_word_lemmatized)
generateWordCloud(freq_lemmatized_word)
```
