Flags

"r" - Read - Default value. Opens a file for reading, error if the file does not exist

"a" - Append - Opens a file for appending, creates the file if it does not exist

"w" - Write - Opens a file for writing, creates the file if it does not exist

"x" - Create - Creates the specified file, returns <mark>an error</mark> if the file exists

### **Counting the occurance of Text


#### Data processing pipelines and iteration


````python
file1 = open("oneline.txt", "r")
onelinedata=file1.read()
file1.close()
onelinedata = onelinedata.strip("\n")
print(onelinedata.split(" "))
````

````python
words=onelinedata.split(" ") # spliting the empty space, put words in the individual list
````

````python
words
````

````python
# Declaring Dictionary type counter
counter ={}
for i in words:
    if i in counter:
        counter[i] += 1
    else:    
        counter[i] =1  # for each new word, we always add 1 (this won't be executed if we already have it in counter)
print(counter)
````

````python
counter['Once']
````

````python
counter['I']
````

````python
# looping into dictionary
for i in counter:
    print(i, counter[i]) # i is the key, counter[key]
````
