#### Counting the occurance of Text

```python
file1 = open("oneline.txt", "r")
onelinedata=file1.read()
file1.close()
onelinedata = onelinedata.strip("\n")
print(onelinedata.split(" "))
words=onelinedata.split(" ")
```

```python
# Declaring Dictionary type counter
counter ={}
for i in words:
    if i in counter:
        counter[i] += 1
    else:
        counter[i] =1
print(counter)
```

### Numpy Array

##### Converting both lists and dictionaries to numpy arrays which gives us the value of the correspondence within the data and map it in a way in which we can be more efficient in our processing.

#### list

```python
# importing library
import numpy as np
 
# initilizing list
lst = [1, 7, 0, 6, 2, 5, 6]
 
# converting list to array
arr = np.array(lst)
#arr2 = np.array(arr)
 
# displaying list
print ("List: ", lst)
 
# displaying array
print ("Array: ", arr)
```

```python
# difference between array and list is that list can have different type value whereas array must only have 1 type
# e.g
new_list = [*lst,'a']
new_array=np.array(new_list)
print(new_list) # the list here is all number except a is a string
print(new_array) # it convert all the number to a string value
```

```python
print(type(lst))
print(type(arr))
```

# Frequency Distribution

```python
import nltk
text=counter
freq = nltk.FreqDist(text)
```

```python
freq
```

```python
for key, val in freq.items():
    print(key + ": " + str(val))
```

```python
dict(freq)==counter # the content produced by nltk and our tokenized string is similar for this case
```
