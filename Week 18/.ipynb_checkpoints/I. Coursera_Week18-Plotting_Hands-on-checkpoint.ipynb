## SNS Distribution Plot  sns.distplot

````python
import seaborn as sns
````

````python
import warnings
#warnings.filterwarnings('ignore')
warnings.filterwarnings('once')
#warnings.filterwarnings('once')
````

````python
plt.figure(figsize=(8,5))
for specie in species:
    data= df.loc[df['species']==specie]
    sns.distplot(data['sepal_length'],
                hist = True,
                kde = True,
                bins=30,
                label = specie,
                kde_kws= {'bw': .1} # kde 0.1 for the bandwidth.
                )

plt.legend()
plt.show()
````

````python
sns.distplot?
````

````python
import warnings
warnings.filterwarnings('ignore')
#warnings.filterwarnings('once')
````

````python
plt.figure(figsize=(8, 5))
for key, dfname in df.groupby('species'):
    sns.distplot(dfname.sepal_length, label=key,
                hist = True,
                kde = True,
                bins=30,
                kde_kws= {'bw': .1} # kde 0.1 for the bandwidth.,
                 
                )

plt.legend()
plt.show()
````

## Newer Library sns.displot without 't'

## Scatter Plot 

````python
plt.figure()

plt.scatter(df.sepal_length, df.sepal_width, s=10)

plt.show()
````

````python
plt.figure()

plt.scatter(df.sepal_length, df.sepal_width, s=10)

plt.show()
````

````python
species_colors= {
    'setosa':      'green',
    'virginica' :  'orange',
    'versicolor' : 'purple'
}
````

````python
plt.figure()

plt.scatter(df.sepal_length, df.sepal_width,
            s=18, 
            c=[species_colors[c] for c in df.species])
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel(' Sepal Length')### CORRECTION!!!!!!!!!!!!!!!!!!
plt.ylabel('Sepal Width') ### CORRECTION!!!!!!!!!!!!!!!!!!
plt.show()
````

````python
species_colors
````

### With Legend

````python
plt.figure()

for specie in species:
    data = df.loc[df.species == specie]
    plt.scatter(data.sepal_length,
               data.sepal_width,
               label=specie,
               s=10,
               c= species_colors[specie])

plt.legend()
plt.show()
````

````python
plt.figure()
size=18
for key, dfname in df.groupby('species'):
    plt.scatter(dfname.sepal_length, dfname.sepal_width, 
                s=size,
               label=key, c=species_colors[key])
plt.legend()
plt.show()
````

````python
plt.figure()
size=18
for key, dfname in df.groupby('species'):
    plt.scatter(dfname.sepal_length, dfname.sepal_width, 
                s=size,
               label=key, c=species_colors[key], alpha=.5)
    size=size*4
plt.legend()
plt.show()
````
