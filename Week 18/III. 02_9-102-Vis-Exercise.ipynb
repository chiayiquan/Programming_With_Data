```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(20,5))


mycolors = {"setosa":'red',"virginica": 'blue', "versicolor":'green'}
i=1
for  key, a in df.groupby('species'):
    ax = fig.add_subplot(1, 3, i, projection='3d')
    
    ax.scatter(a.sepal_length, a.sepal_width, a.petal_length, c=mycolors[key], label=key)
    i=i+1
    plt.legend()

plt.show()
```
