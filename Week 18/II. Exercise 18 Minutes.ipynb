### Create a lineplot similar to the following.  You may choose any colors as long as the pattern looks similar!!!

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('Sales6Country.xlsx')
```

```python
df.shape
```

```python
df.head(4)
```

```python
df.plot.line('Day', 'Y')
```

```python
fig, ax=plt.subplots()
for key, dfname in df.groupby('Group'):
    dfname.plot.line('Day', 'Y', ax=ax, label=key)
plt.show()
```

```python
for key, dfname in df.groupby('Group'):
    plt.plot(dfname.Day, dfname.Y, label=key)
plt.legend()
plt.show()
```
