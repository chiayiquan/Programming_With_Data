### Data Provenance

Data provenance is the documentation of where a piece of data comes from and the processes and methodology by which it was produced. The word provenance originates from the French term 'provenir' meaning 'to come from' and is also known as 'lineage' or 'pedigree'.

https://www.ands.org.au/working-with-data/publishing-and-reusing-data/data-provenance

Data provenance tells how data look like at all stages of the analysis process.

To maintain data provenance:
* Keep a copy of the raw data,
* Keep a log (step/code/instruction) of what kind of cleaning transformations were applied
(with the aim to be `reproducable`) 
* Keep a copy of the cleaned or pristine data.



#### It is good to include Data Provenance when you work outside:

For example, you download/webscrape the data from the website; it may be good to 
- show how to raw data look likes
- how you remove the whitespace
- how you remove columns
- how you format it for machine learning.

I have seen some assignments with good code, but lack of these documentation

It is also important to tell:
* what library you used. 
* whether to install
* warn how long it will take.
* give a backup in case it does not work

#### Good Practice:
When you share data,include intuitive variable names,and try to choose intuitive names such as age_at_diagnosis,as opposed to unintuitive names such as age_DX.

If you join table A, and B.  Try to use the same variable within those different data sets (nric_id, nric_id)

Try to keep different kinds of observations in different tables or different files (covid_positive; covid_negative, good movies; bad movies; good results, bad results).

When you do that, it helps people organize the data in their heads and can possibly analysis one category directly (eg.,only covid positve)

Data should always come with an explanation: 
* what each variable means,
* how the data is coded,
* who to contact with questions
* any information about the licensing or sharing of the data.

Instructions should also be included with your dataset when you share it.

#### Melt, Unmelt, Pivot, Unpivot

```python

import pandas._testing as tm
import numpy as np

def unpivot(frame):
    N, K = frame.shape
    data = {
        "value": frame.to_numpy().ravel("F"),
        "variable": np.asarray(frame.columns).repeat(N),
        "date": np.tile(np.asarray(frame.index), K),
    }
    return pd.DataFrame(data, columns=["date", "variable", "value"])


df = unpivot(tm.makeTimeDataFrame(6))
```
