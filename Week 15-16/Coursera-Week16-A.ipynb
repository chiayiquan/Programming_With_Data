Standardization is to scale the data so the mean of the distribution is zero,and `the standard deviation` is one, also called centering,and the individual re-standardized values are called Z-scores.


The Z-scores represent how far that value is from the average.

This is helpfulbecause it makes comparing variables across different variableswith different unitspossible.The value of these new values are standard deviations.

It's not always easy to interpret what that means,but it does make them comparable when the original units were different.

An indicator can also be standardized. A binary representation of some categorical variable,say someone's nationality.  Even though this is a strange interpretationin terms of a mean and standard deviation,the mathematics works outso that you can use the same standardized variablesin correlations, regressions, and likelihood estimations

Usually used to clean outlier in our data and visualization of different scales:

3 standard deviation above and below the mean

```python
import numpy as np
def centre(v):
    v = (v - np.nanmean(v)) / np.nanstd(v)
    return v
```

```python
from scipy import stats
stats.zscore([0,1,2,3,5,10])
```

```python
centre([0,1,2,3,5,10])
```
